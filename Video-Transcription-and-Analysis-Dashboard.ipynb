```python
import os
import pydub
import tempfile
import speech_recognition as sr
import pandas as pd
import moviepy.editor as mp
from pydub import AudioSegment
from pydub.silence import split_on_silence
```

```python
# getting all list of video files in the /videos directory to use it later
videoFiles = os.listdir('videos')
videoPaths = ['videos/' + filename for filename in videoFiles]
```

```python
# Define the path to the videos folder where all vidoes located
pathToVideos = "videos"

# Define the path to the output folder for extracted audio files
pathToAudio = "audio"

if not os.path.exists(pathToAudio):
    os.makedirs(pathToAudio)
```

```python
# Loop through all the files in the videos folder
for filename in os.listdir(pathToVideos):
    if filename.endswith(".mp4"):
        video = mp.VideoFileClip(os.path.join(pathToVideos, filename))

        # Extract the audio and save it to a file for use it later
        audio = video.audio
        audio.write_audiofile(os.path.join(pathToAudio, f"{filename[:-4]}.wav"), codec='pcm_s16le')

        video.close()
```

```python
#Creating a function below to break down the audio files into small segments for use it later

r = sr.Recognizer()
audiosPath = 'audio'

# Getting all the list of audio files from the audio folder
audioFiles = [f for f in os.listdir(audiosPath) if f.endswith('.wav')]
transcribedTexts = {}

# Loop over all the audio files from audio folder
for audioFile in audioFiles:
    sound = AudioSegment.from_wav(os.path.join(audiosPath, audioFile))
    
    # Split the audio file into segments based on silence
    segments = split_on_silence(sound,
        min_silence_len = 1000,
        silence_thresh = sound.dBFS-14,
        keep_silence=500)
    
    with tempfile.TemporaryDirectory() as temp_dir:
        for i, audioSegment in enumerate(segments, start=1):
            segmentFilename = os.path.join(temp_dir, f"segment{i}.wav")
            audioSegment.export(segmentFilename, format="wav")
            
        text = ""
        for i, audioSegment in enumerate(segments, start=1):
            segmentFilename = os.path.join(temp_dir, f"segment{i}.wav")
            with sr.AudioFile(segmentFilename) as source:
                audioListened = r.record(source)
                try:
                    segmentText = r.recognize_google(audioListened)
                except sr.UnknownValueError:
                    continue
                else:
                    segmentText = f"{segmentText.capitalize()}. "
                    text += segmentText

    # Store the transcribed text in the dictionary for use it later
    transcribedTexts[audioFile] = text
```

```python
import csv

# Saving all the transcribed texts into a CSV file and named it TranscriptionsFile.csv
with open('TranscriptionsFile.csv', mode='w') as file:
    writer = csv.writer(file)
    writer.writerow(['VideoNames', 'Transcriptions'])
    for audioFile, text in transcribedTexts.items():
        writer.writerow([audioFile, text])
```

3. Normalizing the text

```python
# Write your answer/code here
import pandas as pd
import re
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt')

# Define a function to perform text normalization on text
def normalizeTextData(text):
    # Converting here to lowercase
    text = text.lower()
    # Removing non-alphabetic characters here
    text = re.sub(r'[^a-zA-Z\s]', '', text)
    # Tokenizing the text into words here
    words = text.split()
    # Removing stop words here
    words = [word for word in words if word not in stopwords.words('english')]
    # Lemmatizing the words here
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]
    # Join the words back into a string here
    text = ' '.join(words)
    return text
```

```python
# Apply the normalizeTextData function to the 'text_column' column and save the normalized text to a new column called 'NormalizedTranscriptionsData'
alltranscriptions['NormalizedTranscriptionsData'] = alltranscriptions['Transcriptions'].apply(lambda x: normalizeTextData(x))
```

4. Extracting key phrases and topics

```python
# Write your answer/code here
def extractKeyPhrasesTexts(text):
    # Tokenizing the text into individual words here
    tokens = nltk.word_tokenize(text.lower())
    
    # Removing stop words and punctuation here
    stopWords = set(nltk.corpus.stopwords.words('english'))
    tokens = [token for token in tokens if token.isalnum() and token not in stopWords]
    
    # Extracting bigrams and trigrams here
    bigrams = list(nltk.bigrams(tokens))
    trigrams = list(nltk.trigrams(tokens))
    
    # Combining unigrams, bigrams, and trigrams here
    phrases = tokens + [f'{bigram[0]} {bigram[1]}' for bigram in bigrams] + [f'{trigram[0]} {trigram[1]} {trigram[2]}' for trigram in trigrams]
    
    # Count the frequency of each phrase or topic here
    freqDistData = nltk.FreqDist(phrases)
    
    # Return the most common phrases here
    return freqDistData.most_common(10)
```

```python
# Extract key phrases for each transcript here
keyPhrasesData = allNormalizedtranscriptions['NormalizedTranscriptionsData'].apply(extractKeyPhrasesTexts)
```

```python
# Add the key phrases to the dataframe
allNormalizedtranscriptions['keyPhrasesData'] = keyPhrasesData
```
